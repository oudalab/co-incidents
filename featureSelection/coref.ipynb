```python
import json
from pprint import pprint
```

```python

data_dir='/home/yan/eventData/coincident/sorted_all.json'
with open(data_dir) as f:
    content = f.readlines()
#parse the data into a list.
content = [x.strip() for x in content] 
from nltk import sent_tokenize
```

```python
from pymongo import MongoClient
client =MongoClient('mongodb://boomer:pswd@portland.cs.ou.edu:port',maxPoolSize=5)
db = client['lexisnexis']
story=db.disk_stories_full
from bson.objectid import ObjectId
```

```python
# ok use this to test the rank start with 1, so the 1,it mean 0 in the sentence array.
# since there is no yes printed
for i in content:
     if json.loads(i)['id'].split("_")[1]=='0':
            print("yes")
```

```python
# gettting the sentence to the dictionary
count=0
jsoncontent=[]
for count in range(0,len(content)):
        try:
            doc=content[count]
            #index need to minus 1
            index=int(json.loads(doc)['id'].split("_")[1])-1;
            #print(index)            
            docid=json.loads(doc)['id'].split("_")[0]
            wholedoc=story.find_one({"_id":ObjectId(docid)})["article_body"]            
            #parse that body
            #needs to change the str to json object in order to add in another attribute
            jsonele=json.loads(doc)
            #jsonele["sentence"]=sent_tokenize(wholedoc)[index]
            jsonele["doc"]=wholedoc
            jsoncontent.append(jsonele)
            #print(sent_tokenize(wholedoc)[index])
            if(count%500==0):
                print(count)
                #print(jsonele["sentence"])
        except Exception as e:
            #print(e)
            pass
```
